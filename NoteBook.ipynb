## 📖 Data Dictionary

This section provides a detailed breakdown of the features included in the `shopping_trends.csv` dataset.

| Column Name | Description | Data Type |
| :--- | :--- | :--- |
| **Customer ID** | Unique identifier assigned to each customer. | Integer |
| **Age** | The age of the customer. | Integer |
| **Gender** | Gender of the customer (e.g., Male, Female). | Categorical |
| **Item Purchased** | The specific product bought in the transaction. | Categorical |
| **Category** | The broad classification of the item (e.g., Clothing, Accessories). | Categorical |
| **Purchase Amount (USD)** | The total value of the purchase in US Dollars. | Integer |
| **Location** | The US State where the transaction was processed. | Categorical |
| **Size** | The size of the purchased item (e.g., S, M, L, XL). | Categorical |
| **Color** | The color variant of the product. | Categorical |
| **Season** | The specific season of the purchase (Spring, Summer, Fall, Winter). | Categorical |
| **Review Rating** | Customer satisfaction score for the item (1.0 to 5.0). | Float |
| **Subscription Status** | Whether the customer has an active membership (Yes/No). | Boolean |
| **Payment Method** | The method used for payment (e.g., Credit Card, PayPal, Venmo). | Categorical |
| **Shipping Type** | The delivery method selected (e.g., Express, Free Shipping). | Categorical |
| **Discount Applied** | Indicates if a discount was applied to the order (Yes/No). | Boolean |
| **Promo Code Used** | Indicates if a promotional code was used (Yes/No). | Boolean |
| **Previous Purchases** | Total number of prior transactions by the customer. | Integer |
| **Preferred Payment Method** | The customer's top-choice payment option for future use. | Categorical |
| **Frequency of Purchases** | How often the customer shops (e.g., Weekly, Monthly, Bi-Weekly). | Categorical |

## 💎 Feature Engineering
---

```python
# Let's do magic

# Annual Frequency

Frequency_map = {"Fortnightly": 26, "Weekly": 52, "Annually": 1, "Quarterly": 4, "Monthly": 12, "Every 3 Months": 4, "Bi-Weekly": 26}

Data["Frequency of Purchases"] = Data["Frequency of Purchases"].map(Frequency_map, na_action="ignore")
```

```python
# Display Data now

display(Data.sample(5))
```

```python
# Customer Lifetime Value (LTV)

# Projected Annual Revenue  = Frequency * Purchase Amount

Data["Annual Revenue"] = Data["Frequency of Purchases"] * Data["Purchase Amount (USD)"]
```

```python
# Display Data now

display(Data.sample(5))
```

```python
# Binning

# Create Age Groups

Labels = ["Young", "Adult", "Senior"]

Bins = [0, 30, 50, Data["Age"].max()]

Data["Age Group"] = pd.cut(Data["Age"], bins=Bins, labels=Labels)
```

```python
# Display Data now

display(Data.sample(5))
```

```python
# Encoding

Sub_map = {"Yes": 1, "No": 0}

Data["Subscription Status"] = Data["Subscription Status"].map(Sub_map, na_action="ignore")
```

```python
# Display Data now

display(Data.sample(5))
```

## 💰 Business Intelligence Questions
---

```python
# Deal Hunters vs. Loyal

Avg_Previous_Purchases	= Data.groupby("Promo Code Used")["Previous Purchases"].mean()

display(pd.DataFrame(Avg_Previous_Purchases))
```

```python
Conditions = [(Data["Subscription Status"] == 1) | (Data["Previous Purchases"] > 25), (Data["Promo Code Used"] == "Yes") | (Data["Previous Purchases"] < 10)]
Choices = ["Loyal", "Deal Hunter"]

Data["Customer Type"] = np.select(Conditions, Choices, default="Regular")
```

```python
# Display Data now

display(Data.sample(5))
```

```python
Pivot_table = Data.pivot_table(index="Customer Type", values="Purchase Amount (USD)", aggfunc=["count", "mean", "sum"])

Pivot_table.columns = ["Count", "Mean", "Sum"]
Pivot_table = Pivot_table.reset_index()

display(Pivot_table)
```

```python
# The Fashion Trend

Seasonal_Tren = Data.groupby(["Season", "Color"]).size()

Most_Color_per_Season = Seasonal_Tren.groupby(level=0).idxmax().apply(lambda x: x[1])

print(Most_Color_per_Season)
```

```python
most_color_with_count = Seasonal_Tren.groupby(level=0).max().reset_index()
most_color_with_count.columns = ['Season', 'Count']
most_color_with_count['Most_Color'] = Most_Color_per_Season.values

print(most_color_with_count)
```

```python
Pivot_table2 = Data.pivot_table(index="Location",columns="Shipping Type", values="Review Rating", aggfunc=["mean"])

Pivot_table2 = Pivot_table2.reset_index().sort_values(by=('mean', 'Express'), ascending=True)

display(Pivot_table2)
```

```python
# VIP Customers

VIP_THR = Data["Annual Revenue"].quantile(0.9)
VIP_Customers = Data[Data["Annual Revenue"] > VIP_THR]

display(VIP_Customers)
```

```python
VIP_Shipping_Distribution = VIP_Customers["Shipping Type"].value_counts(normalize=True) * 100
print(VIP_Shipping_Distribution)
```

```python
# Age vs. Our Business

age_purchase_total = Data.groupby("Age Group")["Purchase Amount (USD)"].sum().reset_index()
print(age_purchase_total)
```

```python
Top_Customers = Data.sort_values(by="Annual Revenue", ascending=False).reset_index(drop=True)

Top_Customers
```

```python
Top_5_customer = Top_Customers.loc[0:4, ["Customer ID", "Annual Revenue"]]

display(pd.DataFrame(Top_5_customer))
```

```python
# Payments by Age

Payment_by_Age = pd.crosstab(Data["Age Group"], Data["Payment Method"], normalize="index")* 100

display(Payment_by_Age)
```
